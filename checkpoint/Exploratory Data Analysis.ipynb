```python
import matplotlib.pyplot as plt
```

```python
# set global rules
plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)

data.hist(bins=50, figsize=(12, 8)) # generates histograms for each numerical variable
plt.show()
```

The histograms show that many of the variables need to be treated. Some of the variables that are being considered numeric should be factors to be treated as numerical, some variables have outliers that need to be treated, and some variables need to be scaled prior to modeling.

```python
data.plot(kind="scatter", x="LON", y="LAT", grid=True, alpha=0.2) # scatterplot of the long and lat
plt.show()
```

It seems as if there an outlier that is making this scatter plot unhelpful. We need to remove the point at LAT AND LON = 0, so we can zoom in on just the points that fall in LA. Once we are able to do this, we can better examine where current crime hotspots are and potentially overlap a map of the city as well.

```python
plt.matshow(data.corr(numeric_only=True))
continuous_features = data.describe().columns
plt.xticks(range(len(continuous_features)), continuous_features, rotation=45)
plt.yticks(range(len(continuous_features)), continuous_features, rotation=45)
plt.colorbar()
plt.show()
```

The initial correlation matrix highlights what variables are related to one another. We can see that LAT and LON are strongly related, which is intuitive. Crm Cd and Crm Cd2 as well as Crm Cd and Crm Cd1 are also stronly related.
